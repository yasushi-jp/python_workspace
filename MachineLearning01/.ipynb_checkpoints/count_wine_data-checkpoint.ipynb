```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

# ワインデータの読み込み
wine_data = pd.read_csv("data/winequality-white.csv", sep=";", encoding="utf-8")

# 品質ごとにグループ分けして、その数をカウント
count_data = wine_data.groupby('quality')["quality"].count()
print(count_data)

# カウントしたデータをグラフ描画
count_data.plot()
plt.savefig("png/wine-count-plt.png")
plt.show()
```
